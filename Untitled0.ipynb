```python
import pandas as pd
df = pd.read_csv('/content/USArrests.csv')
df.info()
```

```python
df.head()
```

```python

df = df.rename(columns={'Unnamed: 0':'unnamed: 0','Murder':'murder', 'Assault':'assault', 'UrbanPop':'urbanpop', 'Rape':'rape' })
df.head()
```

```python
df['unnamed: 0'].replace(['Alabama', 'Alaska' , 'Arizona' , 'Arkansas' , 'California'], [10,11,12,13,14], inplace=True)
df.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

# Misalkan df adalah data frame Anda, pastikan hanya terdapat nilai numerik di dalamnya

# Drop kolom non-numerik jika ada
df_numerical = df.select_dtypes(include=['float64', 'int64'])

# Menghitung nilai inersia untuk jumlah cluster
inertias = []
for i in range(1, 20):
    km = KMeans(n_clusters=i)
    km.fit(df_numerical)
    inertias.append(km.inertia_)

# Plot nilai inersia
plt.plot(range(1, 20), inertias, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Inertia vs Number of Clusters')
plt.show()

```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
import pandas as pd

# Misalkan df adalah data frame Anda, pastikan hanya terdapat nilai numerik di dalamnya

# Drop kolom non-numerik jika ada
df_numerical = df.select_dtypes(include=['float64', 'int64'])

# Memilih jumlah klaster terbaik
best_k = 5
km = KMeans(n_clusters=best_k)
km.fit(df_numerical)

# Tambahkan kolom label ke dataframe df
df['Label'] = km.labels_

# Buat plot kmeans dengan klaster
plt.figure(figsize=(8, 4))
sns.scatterplot(x=df['urbanpop'], y=df['rape'], hue=df['Label'], palette=sns.color_palette('hls', best_k))
plt.title('KMeans 5 kluster')
plt.xlabel('urbanpop')
plt.ylabel('rape')
plt.show()
```
